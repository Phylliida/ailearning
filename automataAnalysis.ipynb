```python
import cloudpickle
from IPython.display import display
import numpy as np
def analysis():
    f = open("resultRunsWithAutomata.json", "rb")
    runsDict = cloudpickle.load(f)
    f.close() 
    analyzedData = []
    for _, (a, aDat) in runsDict.items():
        avgNumEpochs = np.median([epochs for (epochs, losses) in aDat])+1 # epochs are zero indexed for reasons
        minNumEpochs = np.min([epochs for (epochs, losses) in aDat])+1 # epochs are zero indexed for reasons
        maxNumEpochs = np.max([epochs for (epochs, losses) in aDat])+1 # epochs are zero indexed for reasons
        stdNumEpochs = np.std([epochs for (epochs, losses) in aDat])
        avgLoss0 = np.mean([losses[0] for (epochs, losses) in aDat])
        avgLossWhenFit = np.mean([losses[-1] for (epochs, losses) in aDat])
        analyzedData.append((a, (avgNumEpochs, minNumEpochs, maxNumEpochs, stdNumEpochs, avgLoss0, avgLossWhenFit)))
    dataLabels = ['avg epochs', 'min epochs', 'max epochs', 'std epochs', 'avg loss at epoch 0', 'avg loss when fit']
    def displayData():
        for a, data in analyzedData:
            display(a.toDot())
            print(" ".join([(dataLabels[j] + " " + str(val)) for (j, val) in enumerate(data)]))
    for i, label in enumerate(dataLabels):
        analyzedData.sort(key=lambda x: x[1][i])
        print("\n\n\n\n\n\n\n\nSorted by " + label)
        displayData()
analysis()
```
