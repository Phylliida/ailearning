```python
import stringAutomata
```

```python
# creates a random automata
a = stringAutomata.Automata(nStates=4, symbols=['a', 'b'])
```

```python
# Useful representation for displaying. The octagon is the start state
a.toDot()
```

```python
# tells you the size of the smallest automata that does the same thing, but doesn't modify the automata itself
print(a.complexity())
a.toDot()
```

```python
# modifies the automata to the smallest one that does the same thing
a.minimize()
a.toDot()
```

```python
# generates based on input symbols
# the outputs have 5 outputs because the initial state always outputs a string
inputStr = 'aaba'
inputArr, outputArr = a.generate(4, inputGenerator=lambda i: inputStr[i])
print("inputs:", inputArr)
print("outputs:", outputArr)
```
